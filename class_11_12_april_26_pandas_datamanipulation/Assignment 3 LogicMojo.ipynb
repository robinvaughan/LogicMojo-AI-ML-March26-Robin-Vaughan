Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset("iris")

# Extract feature matrix as NumPy array
# Extract features (exclude the species column as its not numerical in nature)
new_numpy_array_of_features = df.drop(columns='species').to_numpy()
# print(new_numpy_array_of_features)

# Compute: Mean Median Standard deviation Variance (for each feature)
print(df.describe())
print("~" * 60) # Separator line


# Identify: Which feature has highest variability and why it matters in ML.
# The feature with the highest variability is typically a numerical feature with a broad range of values so it matters in ML.
# Here its - petal_length    1.765298
# We had to drop the last column as it is not numerical in nature and hence cannot be used to compute the standard deviation.
print(df.drop(columns='species').std())
print("~" * 60) # Separator line

# Another way to do the same thing.
print(df.iloc[:, 0:-1].std())


# Convert any one feature into shape (n,1) and explain why ML models expect this format

# ML expects input data in a format (n samples (Basically an observation in every single row), 
# 1 feature (One Header Feature for which all the rows are sample observations)) 
# to maintain consistent matrix operations 
# where columns represent features and rows represent observations.
# The new Dataframe is basically of Shape - 150(Rows) by 1(Column).
new_shape_df = df[["petal_length"]]
print(new_shape_df.head())
print(new_shape_df.shape)
```

Q2. Data Selection for Model Input

Context: Models are trained on selective features, not full raw data.

Dataset: Iris

Tasks:

1. Extract:
First 100 samples
Only last 2 features
2. Use boolean masking:
Select samples where petal length is greater than dataset mean
3. Count selected samples
4. Explain how this relates to feature-based filtering in ML

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset("iris")

print(df.head(3))

# Extract: First 100 samples Only last 2 features
print(df.iloc[0:100, -2:])
print("~" * 60) # Separator line

# Use boolean masking: Select samples where petal length is greater than dataset mean for the Column Petal_length.
new_masked_pd = df[df["petal_length"] > df["petal_length"].mean()]
print(new_masked_pd)

# Count selected samples
print(len(new_masked_pd))

# Explain how this relates to feature-based filtering in ML.
# This is a common approach in ML for Data selection and preprocessing as we have selected the samples based on a special condition.
# The condition is our case was fetch only those Rows from the given DF that has petal length value more than the mean/average of Petal length column.
```

Q4. Dataset Understanding (Titanic Dataset)

Context: First step in ML pipeline is dataset inspection.

Dataset: Titanic (Seaborn)

Tasks:

1. Display:
head(), tail()
info(), describe()
2. Identify:
Missing values per column
Numerical vs categorical features
3. Explain:
Why identifying feature types is important before modeling

```python
import seaborn as sns
df = sns.load_dataset('titanic')


# Display: head(), tail() info(), describe()
print(df.head())
print("~" * 60) # Separator line
print(df.tail())
print("~" * 60) # Separator line
print(df.info())
print("~" * 60) # Separator line
print(df.describe())
print("~" * 60) # Separator line


# Identify: Missing values per column Numerical vs categorical features
print(df.isna())
print(df.isnull())
# For a Single Numeric Column - Age.
print(df["age"].isnull().sum())

# Another way is to get all the Columns that are of Type "Float64" or "Int64"
numeric_df = df.select_dtypes(include='number')
print(numeric_df.isnull().sum())

# This will give the result in this format - 
        # survived      0
        # pclass        0
        # age         177
        # sibsp         0
        # parch         0
        # fare          0

# To get all the Columns that are of Type "String" or "Category"
df_categorical = df.select_dtypes(include=['object', 'category'])
print(df_categorical.isnull().sum())


# This will give the result in this format - 
        # sex              0
        # embarked         2
        # class            0
        # who              0
        # deck           688
        # embark_town      2
        # alive            0

# Explain: Why identifying feature types is important before modeling
#  Because it helps to get the Data cleaned, transformed, and understood by machine learning algorithm if we identify the Feature types.
#  Copied from Google - Raw data is often unsuited for direct modeling, and distinguishing feature types ensures proper preprocessing, 
# which directly influences model performance, interpretability, and accuracy. 
```

Q5. Filtering Data for Business Logic

Context: Real ML problems require filtering relevant populations.

Dataset: Titanic

Tasks:

1. Filter:
Female passengers in 1st class
2. Compute:
Survival rate for this group
3. Compare with overall survival rate
4. Interpret:
What insight can be used in ML feature engineering

```python
import seaborn as sns
df = sns.load_dataset('titanic')

print(df.head(3))
print("~" * 60)

# Filter: Female passengers in 1st class
female_passenger_in_first_class = df[(df["sex"] == "female")  & (df["pclass"] == 1)]
print(female_passenger_in_first_class)
print("~" * 60)


#  Compute: Survival rate for this group
#  Basically we need to calculate the percentage for Female passengers in 1st class who survived.
total_number_of_female_first_class_passengers = len(female_passenger_in_first_class)
total_number_of_female_first_class_passengers_who_survived = len(female_passenger_in_first_class[female_passenger_in_first_class["survived"] == 1])

print(total_number_of_female_first_class_passengers)
print(total_number_of_female_first_class_passengers_who_survived)
print("~" * 60)
percentage_for_female_in_first_class_who_survived = (total_number_of_female_first_class_passengers_who_survived/total_number_of_female_first_class_passengers)*100
print(percentage_for_female_in_first_class_who_survived)


#  Compare with overall survival rate
# First we need to calculate overall survival rate or Percentage

# print(df["survived"].sum())
print((len(df[df["survived"] == 1])/len(df))*100)


# Interpret: What insight can be used in ML feature engineering
# Female Survival rate if three times the total population who survived.

```

Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain:
Why different strategies are used for different columns

```python
import seaborn as sns
df = sns.load_dataset('titanic')

print(df.head(3))
print("~" * 60)

# Fill missing age using median
median_of_age_column = df["age"].median()
print(median_of_age_column)
# print("~" * 60)

print(df['age'].isnull().sum())
df['age'] = df['age'].fillna(median_of_age_column)
print(df.head(5))
print(df['age'].isnull().sum())

print("~" * 60)

# Fill missing embarked using mode
print(df['embarked'].isnull().sum())
```
